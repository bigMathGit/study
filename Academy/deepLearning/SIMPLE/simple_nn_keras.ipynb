```python
import tensorflow as tf
import numpy as np
import random
import matplotlib.pyplot as plt
from keras.layers import Dense, Input
from keras.models import Model
```

```python
# y = w * x + b 함수
def myfunc(_x):
  w = 1.3 # 기울기
  b = 2.6 # y 절편.  점(0, 2.6)
  # x 절편은 점(-2, 0)이 됨.
  _y = w * _x + b
  noise = random.random() * 0.1    # Noise
  return _y + noise
```

```python
# random.random() -- (0.0 ~ 1.0)
NUM_DATA = 100      # 데이터 갯수
XVALUE = 5          # X값의 범위 (0.0 ~ 5.0)
# type: python list
xlist = [random.random() * XVALUE for i in range(NUM_DATA)]
xlist.sort()        # sorting
xtest = [random.random() * XVALUE for i in range(NUM_DATA)]
ylist = [myfunc(x) for x in xlist]
print('xlist :',xlist)
print('ylist :',ylist)
print('xlist type :', type(xlist))
print('ylist type :', type(ylist))
```

```python
# actual data 산점도
fig, ax = plt.subplots()
ax.plot(xlist, ylist, label='actual')
plt.show()
```

```python
# type 변환 python list -> numpy ndarray
# type: numpy ndarray
xlist = np.array(xlist)
ylist = np.array(ylist)
print('xlist.shape :',xlist.shape)  # shape ==  (10,)
print('ylist.shape :',ylist.shape)  # shape ==  (10,)
```

```python
# type 에 맞추기 위해 reshape
xlist = xlist.reshape((NUM_DATA, 1))  # shape ==  (10,1)
ylist = ylist.reshape((NUM_DATA, 1))  # shape ==  (10,1)
print('xlist.shape :', xlist.shape)
print('ylist.shape :', ylist.shape)
```

```python
# keras
inputs = Input(shape=(1,))
# activation -> https://keras.io/activations/
preds = Dense(1, activation='linear')(inputs)

# keras 모델 구성하기
model = Model(inputs=inputs, outputs=preds)

# 모델 학습과정 설정하기
# optimizer -> https://keras.io/optimizers/
# loss      -> https://keras.io/losses/
model.compile(optimizer='adam', loss='mean_squared_error')
# model 학습
hist = model.fit(xlist, ylist, epochs=2000)
# model 예상
predict = model.predict(xtest)
```

```python
# plt
fig, loss_ax = plt.subplots()

# loss_ax.plot(hist.history['loss'], label='train loss')
loss_ax.scatter(xlist, ylist, label='actual')
loss_ax.scatter(xtest, predict, label='test')

loss_ax.set_xlabel('X')
loss_ax.set_ylabel('Y')

loss_ax.legend()

plt.show()
```
